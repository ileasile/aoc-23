```kotlin
fun plotFuns(xs: List<Int>, fns: List<(Int) -> Long>) =
    plot {
        val n = fns.size
        line {
            x(xs.repeatList(n))
            y(fns.flatMap { f -> xs.map(f) })
            color((1..n).flatMap { i -> List(xs.size) { i.toString() } })
        }
    }
```

```kotlin
val valsRange = 0..50
val funsRange = 0..50
plotFuns(valsRange.toList(), funsRange.map { i -> {j -> Ag(i, j)} })
```

```kotlin
plotFuns(valsRange.toList(), funsRange.map { i -> {j -> if (i >= j) Cn(i, j) else 0} })
```
